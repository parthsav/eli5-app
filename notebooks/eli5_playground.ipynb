# ELI5 Playground

Quick notebook to prototype the **Explain Like I'm 5** prompt against the OpenAI API before wiring it into the FastAPI backend.

Run the cells top to bottom. Make sure `OPENAI_API_KEY` is set in a `.env` file at the project root (or exported in your shell).

```python
import os
from pathlib import Path
from dotenv import load_dotenv

load_dotenv(Path('..') / '.env')
assert os.getenv('OPENAI_API_KEY'), 'OPENAI_API_KEY not set. Add it to .env at the project root.'
print('Key loaded:', os.getenv('OPENAI_API_KEY')[:7] + '...')
```

```python
from openai import OpenAI

client = OpenAI()
MODEL = 'gpt-4o-mini'

SYSTEM_PROMPT = (
    "You are 'ELI5 Bot'. Explain whatever the user asks as if they are 5 years old. "
    "Use simple words, short sentences, a friendly tone, and one concrete analogy. "
    "Keep responses under 150 words."
)

def explain_like_im_five(question: str) -> str:
    response = client.chat.completions.create(
        model=MODEL,
        messages=[
            {'role': 'system', 'content': SYSTEM_PROMPT},
            {'role': 'user', 'content': question},
        ],
    )
    return response.choices[0].message.content
```

```python
questions = [
    'What is gravity?',
    'What is recursion?',
    'Why is the sky blue?',
]

for q in questions:
    print('Q:', q)
    print('A:', explain_like_im_five(q))
    print('-' * 60)
```
